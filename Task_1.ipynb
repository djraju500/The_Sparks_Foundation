## TASK-1:- OBJECT DETECTION-OpenCV

```python
import cv2
import matplotlib.pyplot as plt
```

```python
config_file = r'C:\Users\DJ RAJU\Desktop\GRIP FOUNDATION\ObjectDetector-OpenCV-main\ObjectDetector-OpenCV-main\ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'

frozen_model = r'C:\Users\DJ RAJU\Desktop\GRIP FOUNDATION\ObjectDetector-OpenCV-main\ObjectDetector-OpenCV-main\frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
classlabel =[]
file_name = r'C:\Users\DJ RAJU\Desktop\GRIP FOUNDATION\ObjectDetector-OpenCV-main\ObjectDetector-OpenCV-main\coco_label.txt'
with open(file_name,'rt') as ftp:
    classlabel = ftp.read().rstrip('\n').split('\n')
```

```python
print(classlabel)
```

```python
print(len(classlabel))
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)          #255/127.5
model.setInputMean((127.5,127.5,127.5))   # mobilenet=>[1,-1]
model.setInputSwapRB(True)
```

## Read an Image

```python
img = cv2.imread(r'C:\Users\DJ RAJU\Desktop\GRIP FOUNDATION\ObjectDetector-OpenCV-main\ObjectDetector-OpenCV-main\image.jpg')
```

```python
plt.imshow(img)#bgr
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
ClassIndex, confidence, bbox = model.detect(img,confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
    cv2.rectangle(img,boxes,(255,0,0),2)
    cv2.putText(img,classlabel[ClassInd-1],(boxes[0]+10,boxes[0]+40),font,fontScale=font_scale,color=(0,255,0), thickness=3)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

## Video Demo

```python
cap = cv2.VideoCapture(1)

#check if the video is opened correctly

if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError('Cannot open video')
    
    
font_scale=3
font= cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()
    
    ClassIndex,confidence,bbox = model.detect(frame,confThreshold=0.50)
    
    print(ClassIndex)
    if (len(ClassIndex)!=0):
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if (ClassInd<=80):
                cv2.rectangle(frame,boxes,(255,0,0),2)
                cv2.putText(frame,classlabel[ClassInd-1],(boxes[0]+10,boxes[0]+40),font,fontScale=font_scale,color=(0,255,0), thickness=3)
    cv2.imshow('Object Detection Tutorial',frame)
    
    if cv2.waitKey(2) & 0XFF == ord('q'):
        break
cap.release()
cv2.destroyWindow()
                
```
